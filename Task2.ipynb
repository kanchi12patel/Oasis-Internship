```python
import pandas as pd
import matplotlib.pyplot as plt

# Load unemployment rate data
unemployment_data = pd.read_csv('Unemployment in India.csv')

# Remove leading space from column name
unemployment_data.rename(columns={' Date': 'Date'}, inplace=True)

# Remove leading spaces from 'Date' column
unemployment_data['Date'] = unemployment_data['Date'].str.strip()

# Convert 'Date' column to datetime format
unemployment_data['Date'] = pd.to_datetime(unemployment_data['Date'], format='%d-%m-%Y')

# Plot the unemployment rate over time
plt.figure(figsize=(10, 6))
plt.plot(unemployment_data['Date'], unemployment_data[' Estimated Unemployment Rate (%)'])
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


```

```python
import pandas as pd

# Load the dataset
unemployment_data = pd.read_csv('Unemployment in India.csv')

# Display column names
print("Column Names:", unemployment_data.columns)
```

```python
mean_unemployment = unemployment_data[' Estimated Unemployment Rate (%)'].mean()
median_unemployment = unemployment_data[' Estimated Unemployment Rate (%)'].median()
std_unemployment = unemployment_data[' Estimated Unemployment Rate (%)'].std()
print("Mean Estimated Unemployment Rate:", mean_unemployment)
print("Median Estimated Unemployment Rate:", median_unemployment)
print("Standard Deviation of Estimated Unemployment Rate:", std_unemployment)
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.histplot(unemployment_data[' Estimated Unemployment Rate (%)'], bins=20, kde=True)
plt.title('Distribution of Estimated Unemployment Rate')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 4: Hypothesis Testing

from scipy import stats

# Example: Test if there is a significant difference in unemployment rates before and during the pandemic
before_pandemic_data = unemployment_data[unemployment_data[' Date'] < '2020-01-01'][' Estimated Unemployment Rate (%)']
during_pandemic_data = unemployment_data[unemployment_data[' Date'] >= '2020-01-01'][' Estimated Unemployment Rate (%)']

t_stat, p_value = stats.ttest_ind(before_pandemic_data, during_pandemic_data)

print('T-statistic:', t_stat)
print('P-value:', p_value)

if p_value < 0.05:
    print('There is a significant difference in unemployment rates before and during the pandemic.')
else:
    print('There is no significant difference in unemployment rates before and during the pandemic.')
```

```python
from scipy import stats

# Example: Test if there is a significant difference in unemployment rates across regions
regions = unemployment_data['Region'].unique()

# Perform pairwise t-tests between unemployment rates of different regions
for i in range(len(regions)):
    for j in range(i+1, len(regions)):
        region1_data = unemployment_data[unemployment_data['Region'] == regions[i]][' Estimated Unemployment Rate (%)']
        region2_data = unemployment_data[unemployment_data['Region'] == regions[j]][' Estimated Unemployment Rate (%)']

        t_stat, p_value = stats.ttest_ind(region1_data, region2_data)

        print(f"T-test between {regions[i]} and {regions[j]}:")
        print('T-statistic:', t_stat)
        print('P-value:', p_value)

        if p_value < 0.05:
            print('There is a significant difference in unemployment rates between', regions[i], 'and', regions[j])
        else:
            print('There is no significant difference in unemployment rates between', regions[i], 'and', regions[j])
        print()
```
